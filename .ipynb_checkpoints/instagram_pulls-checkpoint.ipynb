## Useful Hashtags 

We can use the `get_related_hashtag_volume` method to get hashtags related to given base words or hashtags that we're interested in. This method returns the suggested search results from inputing text to the instagram search bar. 

```python
# instantiate the class 
scraper = instagram_scraper.instagram_scraper() 

# define the hashtags you 
hashtags = ['sunset',                                  
            'sky',
            'rainbow',
            'sandiego',
            'california',
            'pacific',
            'cloud',
            'hidef',
            'pixel']

data = scraper.get_related_hashtag_volume(hashtags)
data = data.drop_duplicates(subset='hashtags')
data = data.sort_values('posts', ascending=False)
```

### *Copy & Paste Hashtags for Your Post* 

We can print out the hashtags so you can copy and paste them into your posts. Below we are only keeping hastags that have over a million posts. And since ***Instagram only lets you post a maximum of 30 hashtags*** we limit the print out to 30. 

```python
'#'+' #'.join(data[data['posts']>1000000]['hashtags'].values[:30])
```
